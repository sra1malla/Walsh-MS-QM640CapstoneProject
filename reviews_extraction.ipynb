```python
import pandas as pd
import numpy as np
import traceback, logging
import datetime
import requests
import json
```

```python
def extract_apps(lookup_file_path, store):
    try:
        app_lookup = pd.read_csv(lookup_file_path)
        apps = app_lookup[app_lookup['store'] == store]
        return apps
    except Exception as e:
        logging.error(traceback.format_exc())
        raise e
```

## Google Play Store

```python
lookup_file_path = './lookup/CompetitorApps_Lookup.csv'
store = 'Google Play Store'

gplay_apps = extract_apps(lookup_file_path, store)
gplay_apps
```

```python
df_gplay_reviews = pd.DataFrame()

# Review Extraction
gplay_apps = extract_apps(lookup_file_path, store)
for i in range(gplay_apps.shape[0]):
    store = gplay_apps.iloc[i, 0]
    app_name = gplay_apps.iloc[i, 1]
    app_id = gplay_apps.iloc[i, 3]
    print("Initiating reviews extraction for", store, " - ", app_name)
    total_pages = -1
    for i in range(1, 100):
#         print(i, total_pages)
        if i == total_pages + 1:
            break
        else:
            from_date = "2020-01-01"
            to_date = "2025-06-30"
            page = i
            url = f"https://api.appfollow.io/api/v2/reviews?ext_id={app_id}&from={from_date}&to={to_date}&page={page}"

            headers = {
                "accept": "application/json",
                "X-AppFollow-API-Token": "eyJ0eXAiOiJKV1QiLCJhbGciOiJIUzI1NiJ9.eyJpc3MiOiJhcGl2MiIsInN1YiI6MjA3Nzc4LCJqdGkiOjM3NjQ4LCJzY29wZXMiOiJkcncifQ.PH5j33bFiQuc6Usc0rNYz2lpgFhUOmHBQ38napG8RvU"
            }

            response = requests.get(url, headers=headers)
            response.encoding = 'UTF-8'
            data = json.loads(response.text)
            if data['reviews']['total'] > 0:
                df = pd.json_normalize(data['reviews']['list'])
                total_pages = data['reviews']['page']['total']
            else:
                total_pages = 0
                print("No reviews available for", app_name)
                break

            if df.shape[0] > 0:
                df['store'] = store
                df['app_name'] = app_name
                df_gplay_reviews = pd.concat([df_gplay_reviews, df], ignore_index=True)
                df_gplay_reviews.reset_index(inplace=True, drop=True)
                print("Success - Appended ", df.shape[0], "reviews for", app_name)
            else:
                print("No reviews available for", app_name)
```

## Merge Reviews

```python
def merge_df(df1, df2):
    if np.array_equiv(df1.columns, df2.columns):
        df = pd.concat([df1, df2], axis=0)
        df.reset_index(inplace=True, drop=True)
        return df
    else:
        print("Error: Merge Failed! Check Dataframes and Columns")
```

```python
df_merged = merge_df(df_gplay_prcsd, df_appstr_prcsd)
print("Total Reviews Merged:", df_merged.shape)

print("No Of Duplicates:", df_merged.duplicated(
    subset=['reviewId', 'userName', 'reviewContent', 'rating', 'reviewedAt', 'store']
).sum())

df_merged['res_ind'] = 0
df_merged.loc[df_merged.replyContent.isna(), 'res_ind'] = 1

df_merged['res_ind_rnk'] = df_merged.groupby(
    ['reviewId', 'userName', 'reviewContent', 'rating', 'reviewedAt', 'store'],
    dropna=False
)['res_ind'].transform('rank', method='first', ascending=True)

df_merged = df_merged[df_merged.res_ind_rnk == 1].copy().drop(columns=['res_ind_rnk', 'res_ind'])

print("Total Unique Reviews Merged:", df_merged.shape)
```

```python
from scipy.stats import norm
import math

# Parameters
alpha = 0.05
beta = 0.20
p1 = 0.80
p2 = 0.70
p_bar = (p1 + p2) / 2

z_alpha = norm.ppf(1 - alpha / 2)
z_beta = norm.ppf(1 - beta)

numerator = 2 * (z_alpha + z_beta)**2 * p_bar * (1 - p_bar)
denominator = (p1 - p2)**2

n_per_group = math.ceil(numerator / denominator)
n_total = 2 * n_per_group

print("Required per group:", n_per_group)
print("Total required:", n_total)
```
